```python
import numpy as np
from scipy.stats import boxcox, yeojohnson, shapiro

test = final_churn_Df.copy()

def log_transformation(df,col):
    df['log_trans'] = np.log1p(df[col])
    
def sqrt_transformation(df,col):
    df['sqrt_trans'] = np.sqrt(df[col])
    
def boxcox_transformation(df, col):
    # Box-Cox transformation requires strictly positive values
    positive_values = df[col] + 1  # Ensure all values > 0
    df['box_trans'], lambda_value = boxcox(positive_values)
    print(f"Box-Cox Lambda for column '{col}': {lambda_value}")
    
def yeojohnson_transformation(df, col):
    # Yeo-Johnson transformation handles both positive and negative values
    df['John_trans'], lambda_value = yeojohnson(df[col])
    print(f"Yeo-Johnson Lambda for column '{col}': {lambda_value}")    

# Applying transformations
log_transformation(test, 'recency')
sqrt_transformation(test, 'recency')
boxcox_transformation(test, 'recency')
yeojohnson_transformation(test, 'recency')
```

```python
import pandas as pd
from sklearn.feature_selection import chi2
from scipy.stats import chi2_contingency
import numpy as np

# Assuming `final_churn_Df` is the DataFrame with one-hot encoded columns

# Separate target and features
X = final_churn_Df.drop(columns=['Churn'])
y = final_churn_Df['Churn']

# Select categorical columns after one-hot encoding
categorical_cols = [
    'method_name_international', 'method_name_priority', 'method_name_standard', 'method_name_unknown',
    'language_name_aleut', 'language_name_arabic', 'language_name_british english', 'language_name_canadian english',
    'language_name_chinese', 'language_name_english', 'language_name_french', 'language_name_gaelic',
    'language_name_galician', 'language_name_german', 'language_name_greek', 'language_name_italian',
    'language_name_japanese', 'language_name_latin', 'language_name_malaysian', 'language_name_middle english',
    'language_name_multiple languages', 'language_name_norwegian', 'language_name_portuguese', 'language_name_russian',
    'language_name_serbian', 'language_name_spanish', 'language_name_swedish', 'language_name_united states english',
    'language_name_welsh', 'order_month_name_april', 'order_month_name_august', 'order_month_name_december',
    'order_month_name_february', 'order_month_name_january', 'order_month_name_july', 'order_month_name_june',
    'order_month_name_march', 'order_month_name_may', 'order_month_name_november', 'order_month_name_october',
    'order_month_name_september', 'status_value_cancelled', 'status_value_delivered', 'status_value_delivery in progress',
    'status_value_order received', 'status_value_pending delivery', 'status_value_returned'
]

significant = []
non_significant = []

# Loop through each categorical column and perform the Chi-Square test
for column in categorical_cols:
    # Create contingency table between the categorical column and target variable
    contingency_table = pd.crosstab(X[column], y)
    
    # Perform Chi-Square Test
    chi2_stat, p_val, _, _ = chi2_contingency(contingency_table)
    if p_val < 0.05:
        significant.append(column)
    else:
        non_significant.append(column)
print(significant)
print(non_significant)
```
